```python
# problmem 4
# A three digit number is said to be an 'Armstrong number' if
# the sum of the third power of its individual digits is 
# equal to the number itself. write a program to check whether
# a number is armstrong or not.

num = int(input("enter the number :"))
sum = 0
temp = num
print(temp)
while int(temp)>0 :
    rem = int(temp) % 10
    sum += pow(rem,3)
    temp /=10
if num == sum :
    print("Armstrong")
else:
    print("Not Armstrong")
```
