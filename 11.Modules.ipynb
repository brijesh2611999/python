# Modules
1. Same as a code library.
2. A file containing a set of function you want to include in your application.

1. math module
2. random module

```python
help('modules')
```

```python
import math
print(math.pi)
print(math.e)
print(math.factorial(5))
print(math.floor(4.3))
print(math.ceil(4.3))
print(math.sqrt(25))
```

```python
import random
print(random.randint(1,100))
list = [1,2,3,4,5]
random.shuffle(list)
print(list)
```

```python
import time
print(time.time())
print(time.ctime())
print(time.sleep(2))
```

```python
import os
print(os.getcwd())
print(os.listdir())
```
